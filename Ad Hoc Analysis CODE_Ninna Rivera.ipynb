```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json("transaction-data-adhoc-analysis.json")
```

```python
df["transaction_items"]=df["transaction_items"].str.split(";")
df= df.explode("transaction_items")
```

```python
df["transaction_items_list"]=df["transaction_items"].str.split(",")
df[["brand", "item", "quantity"]] = pd.DataFrame(df.transaction_items_list.tolist(), index= df.index)
df["quantity"]=df["quantity"].str.extract('(\d+)')
df["transaction_month"] = pd.to_datetime(df["transaction_date"]).dt.month
df["item"].unique()
```

```python
#this only counts the number of rows, not the total quantity per item per month
df.groupby(["transaction_month","item"]).agg({"quantity":"count"})
```

```python
df["quantity"] = df["quantity"].astype(int)
df.groupby(["transaction_month","item"]).agg({"quantity":"sum"})
```

```python
#replace transaction_month number labels into text
df["transaction_month"] = df["transaction_month"].replace(
    {1:"January",2:"February",3:"March",4:"April",5:"May",6:"June"})
```

## ---Start of Code for the Table of Customer Metrics---

```python
df3 = pd.DataFrame(index=["Repeaters", "Inactive", "Engaged"], columns=["January", "February", "March", "April", "May", "June"])
```

```python
df3
```

```python
df4 =df.groupby("username").agg({"transaction_month": "unique"})
df4
```

```python
df4.explode("transaction_month")
```

```python
df5 = pd.get_dummies(df4.explode("transaction_month"))
df5
```

```python
df5= df5.groupby("username").sum()
df5
```

```python
df5["transaction_month_January"]= df5["transaction_month_January"].astype(bool)
df5["transaction_month_February"]= df5["transaction_month_February"].astype(bool)
df5["transaction_month_March"]= df5["transaction_month_March"].astype(bool)
df5["transaction_month_April"]= df5["transaction_month_April"].astype(bool)
df5["transaction_month_May"]= df5["transaction_month_May"].astype(bool)
df5["transaction_month_June"]= df5["transaction_month_June"].astype(bool)
```

```python
months = ["January", "February", "March", "April", "May", "June"]

df5["february_repeater"] = df5["transaction_month_January"] & df5["transaction_month_February"]
for i in range(2, 6):
    new_column_name = months[i].lower() + "_repeater"
    repeater_column = months[i - 1].lower() + "_repeater"
    transaction_column = "transaction_month_" + months[i]
    df5[new_column_name] = df5[repeater_column] & df5[transaction_column]
```

```python
df5.head()
```

## REPEATERS: Customers from the current month who also purchased in the previous month

```python
df3.at["Repeaters", "January"] = 0
for i in range(1, 6):
    repeaters_column = months[i]
    repeaters_column_value = months[i].lower() + "_repeater"
    print (repeaters_column, repeaters_column_value)
    df3.at["Repeaters",repeaters_column] = sum(df5[repeaters_column_value])
```

```python
df3
```

## INACTIVE: Customers who have purchase history but do not have a purchase for the current month

```python
df5["february_inactive"] = (df5["transaction_month_January"] & (~ df5["transaction_month_February"]))
df5["march_inactive"] = (df5["transaction_month_January"]|df5["transaction_month_February"]) & (~df5["transaction_month_March"])
df5["april_inactive"] = (df5["transaction_month_January"]|df5["transaction_month_February"]|df5["transaction_month_March"]) & (~df5["transaction_month_April"])
df5["may_inactive"] = (df5["transaction_month_January"]|df5["transaction_month_February"]|df5["transaction_month_March"]|df5["transaction_month_April"]) & (~df5["transaction_month_May"])
df5["june_inactive"] = (df5["transaction_month_January"]|df5["transaction_month_February"]|df5["transaction_month_March"]|df5["transaction_month_April"]|df5["transaction_month_May"]) & (~df5["transaction_month_June"])
```

```python
df5.head(5)
```

```python
df3.at["Inactive", "January"] = 0
df3.at["Inactive", "February"] = sum(df5["february_inactive"])
df3.at["Inactive", "March"] = sum(df5["march_inactive"])
df3.at["Inactive", "April"] = sum(df5["april_inactive"])
df3.at["Inactive", "May"] = sum(df5["may_inactive"])
df3.at["Inactive", "June"] = sum(df5["june_inactive"])
df3
```

## ENGAGED: Customers who have consistently purchased every single month

```python
df5["february_engaged"] = (df5["transaction_month_January"] & df5["transaction_month_February"])
df5["march_engaged"] = (df5["transaction_month_January"] & df5["transaction_month_February"] & df5["transaction_month_March"])
df5["april_engaged"] = (df5["transaction_month_January"] & df5["transaction_month_February"] & df5["transaction_month_March"] & df5["transaction_month_April"])
df5["may_engaged"] = (df5["transaction_month_January"]& df5["transaction_month_February"] & df5["transaction_month_March"] & df5["transaction_month_April"] & df5["transaction_month_May"])
df5["june_engaged"] = (df5["transaction_month_January"]& df5["transaction_month_February"] & df5["transaction_month_March"]& df5["transaction_month_April"]& df5["transaction_month_May"] & df5["transaction_month_June"])
```

```python
df5.head(5)
```

```python
df3.at["Engaged", "January"] = sum(df5["transaction_month_January"])
df3.at["Engaged", "February"] = sum(df5["february_engaged"])
df3.at["Engaged", "March"] = sum(df5["march_engaged"])
df3.at["Engaged", "April"] = sum(df5["april_engaged"])
df3.at["Engaged", "May"] = sum(df5["may_engaged"])
df3.at["Engaged", "June"] = sum(df5["june_engaged"])
df3
```
